```python
import pandas
import numpy
import matplotlib.pyplot as plt
```

```python
data = {
    'Produto': ['A', 'B', 'C', 'D'],
    'Vendas': [100, 200, 200,300]
}

df = pandas.DataFrame(data)

print(df)
```

```python
df.plot(kind='bar', x='Produto', y='Vendas')
plt.title('Venda de Produto')
plt.xlabel('Produto')
plt.ylabel('Venda')
plt.show()
```
